```python
import pickle
from PIL import Image
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
```

```python
file_path = './datasets'
file_suffix = '.pkl'
test_benign_local_path = f'{file_path}/test_benign{file_suffix}'

# Load the pickle file from local
with open(test_benign_local_path, 'rb') as f:
    test_benign = pickle.load(f)
```

```python
test_malignant_local_path = f'{file_path}/test_malignant{file_suffix}'

# Load the pickle file from local
with open(test_malignant_local_path, 'rb') as f:
    test_malignant = pickle.load(f)
```

```python
train_malignant_local_path = f'{file_path}/train_malignant{file_suffix}'

# Load the pickle file from local
with open(train_malignant_local_path, 'rb') as f:
    train_malignant = pickle.load(f)
```

```python
train_benign_local_path = f'{file_path}/train_benign{file_suffix}'

# Load the pickle file from local
with open(train_benign_local_path, 'rb') as f:
    train_benign = pickle.load(f)
```

```python
# Combine train benign & malignant files
X = train_benign + train_malignant
# Convert values into numpy arrays
X = np.array(X)
# Create y_train as list from benign & malignant files
y = [0] * len(train_benign) + [1] * len(train_malignant)
# Convert values into numpy arrays

#y = ["0"] * len(train_benign) + ["1"] * len(train_malignant)
#len(y)

# Combine test benign & malignant files
# Create y_test from benign & malignant files
X_test = test_benign + test_malignant
X_test = np.array(X_test)

y_test = [0] * len(test_benign) + [1] * len(test_malignant)


```

```python
y = np.array(y)
y_test = np.array(y_test)
```

```python
# fine tuning with Grid Search
def build_classifier(optimizer):
    # first step: create a Sequential object, as a sequence of layers. B/C NN is a sequence of layers.
    nn_model = Sequential()
    # add the first hidden layer
    nn_model.add(Dense(units=5,kernel_initializer='glorot_uniform',
                    activation = 'relu'))
    # add the second hidden layer
    nn_model.add(Dense(units=5,kernel_initializer='glorot_uniform',
                    activation = 'relu'))
    # add the output layer
    nn_model.add(Dense(units=1,kernel_initializer='glorot_uniform',
                    activation = 'sigmoid'))
    # compiling the NN
    nn_model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
    return nn_model

new_classifier = KerasClassifier(build_fn=build_classifier)

# create a dictionary of hyper-parameters to optimize
parameters = {'batch_size':[25,32], 'nb_epoch':[1,2,10],'optimizer':['adam','rmsprop','sgd']}
grid_search = GridSearchCV(estimator = new_classifier, param_grid = parameters, scoring = 'accuracy', cv=10)
grid_search = grid_search.fit(X[int(0.5*len(X)):],y[int(0.5*len(X)):])
# decreased train size for faster results. This is one way to get results faster if runing short on time
best_parameters = grid_search.best_params_ 
best_accuracy = grid_search.best_score_
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
```

```python
# Define the model
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2, activation='sigmoid')
])
model
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
batch_size = 32
epochs = 10
history = model.fit(
    X, y,
    validation_data=(X_test, y_test),
    epochs=epochs
)
```
