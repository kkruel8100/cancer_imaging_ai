```python
import pickle
import requests
import io
import os
from dotenv import load_dotenv
import requests
from PIL import Image
```

```python
# Get environment variables
env_path = '.env'
load_dotenv(dotenv_path=env_path)

test_benign_path = os.getenv("test_benign")
test_malignant_path = os.getenv("test_malignant")
train_benign_path = os.getenv("train_benign")
train_malignant_path = os.getenv("train_malignant")
```

```python
response_test_benign = requests.get(test_benign_path)

print("Status Code:", response_test_benign.status_code)
```

```python
test_benign_images = pickle.load(io.BytesIO(response_test_benign.content))
first_test_benign = test_benign_images[0]
image = Image.fromarray(first_test_benign)
display(image)
```

```python
file_path = './datasets'
file_suffix = '.pkl'
```

```python
# Save the images to a local file
test_benign_local_path = f"{file_path}/test_benign{file_suffix}";

with open(test_benign_local_path, 'wb') as file:
    # Use pickle.dump to store the list of images
    pickle.dump(test_benign_images, file)
```
